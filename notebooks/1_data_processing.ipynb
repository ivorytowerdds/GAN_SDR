```python
df['whole'] = pd.Series(df.values.tolist()).str.join('')
```

```python
with open('encoded.txt', "wt") as fout:
    count = 0
    sentence = []
    for i in df['whole']:
        if count < 30:
            sentence.append(i)
            count += 1
        else:
            sen = ' '.join(_ for _ in sentence)
            fout.write(sen)
            fout.write('\n')
            count = 0
            sentence = []
```

```python
import re
with open('encoded.txt', "rt") as fin:
    with open('int_raw.txt', "wt") as fout:
        line = fin.read()
        int2word = dict(enumerate(set(line.split())))
        word2int = {v: str(k) for k, v in int2word.items()}
        for x in word2int:
            line = re.sub(x, word2int[x], line)
        fout.write(line)
```

```python
with open('encoded.txt', "rt") as fin:
    with open('word.vocab', "wt") as fout:
        line = fin.read()
        line = list(set(line.split()))
        for word in line:
            fout.write(word)
            fout.write('\n')
```
