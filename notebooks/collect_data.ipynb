```python
import requests
import pandas as pd
import datetime
t = pd.date_range(start = '1/1/2014', end='2/27/2019', freq='D')
t=t.strftime('%Y_%m_%d')

#https://kgc0418-tdw-data2-0.s3.amazonaws.com/slices/CUMULATIVE_CREDITS_2018_01_01.zip
urlh = 'https://kgc0418-tdw-data2-0.s3.amazonaws.com/slices/CUMULATIVE_CREDITS_'
urlt='.zip'
for i in t:
    url = urlh+i+urlt
    #print(url)
    r = requests.get(url)
    if r.status_code != 200:
        print(i + " file is missing")
    else:
        with open(i+urlt, "wb") as code:
            code.write(r.content)
        
```

```python
import zipfile
import os

file_list = os.listdir()
for file_name in file_list:
    if os.path.splitext(file_name)[1] == '.zip':
        print (file_name)
        file_zip = zipfile.ZipFile(file_name, 'r')
        for file in file_zip.namelist():
            try:
                outpath = 'csv_file'
                file = file_zip.extract(file, outpath)
                new_file_name = "csv_file/" + os.path.splitext(file_name)[0] +'.csv'
                os.rename(file, new_file_name)
            except KeyError:
                {}
        file_zip.close()
        os.remove(file_name)
```

```python
import os
filenames = os.listdir('csv_file')

file=open('all.csv','w')

for (i, filename) in enumerate(filenames):
    if filename != ".DS_Store":
        filepath='csv_file/'+filename
        f = open(filepath, encoding="utf8", errors='ignore')
        if i != 0:
            next(f)
        for line in f:
            file.writelines(line)
        file.write(' ')
file.close()
```

```python
data1 = data1[data1['CLEARED'] == 'U']
data1.info()
```

```python
data2 = data1[data1['TAXONOMY'] == ("Credit:Index:CDX:CDXIG" or "Credit:Index:CDX:CDXHY")]
```

```python
data2.reset_index(inplace = True)
```

```python
data2['EFFECTIVE_DATE'] = pd.to_datetime(data2['EFFECTIVE_DATE'])
```

```python
data2['END_DATE'] = pd.to_datetime(data2['END_DATE'])
```

```python
interval = []
for i in range(data2.shape[0]):
    diff = data2['END_DATE'][i] - data2['EFFECTIVE_DATE'][i]
    if diff.days <= 1980 and diff.days >= 1620:
        interval.append(True)
    else:
        interval.append(False)
```

```python
data2['interval'] = interval
```

```python
data2.rename(columns = {'interval':'INTERVAL'}, inplace = True)
```

```python
data3 = data2[data2['INTERVAL']]
```
